```python
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
import seaborn as sns
import os

def process_spectro_data():
    """
    Main program: Process spectroelectrochemical data and generate publication-quality figures and tables.
    """
    
    # --- 1. Style Configuration ---
    sns.set_theme(style="ticks", context="notebook")
    plt.rcParams.update({
        'font.family': 'DejaVu Sans',
        'axes.unicode_minus': False,
        'axes.linewidth': 1.25,
        'axes.edgecolor': '#1f2933',
        'axes.facecolor': '#f9fafb',
        'figure.facecolor': '#fcfcfc',
        'grid.color': '#d2d6dc',
        'grid.alpha': 0.35,
        'grid.linestyle': '--',
        'grid.linewidth': 0.8,
        'legend.frameon': True
    })

    # --- 2. File and Parameter Setup ---
    files = {
        '+0.4': '04v.csv',
        '+0.2': '02v.csv',
        '+0.1': '01v.csv',
        '0.0': '0v.csv',
        '-0.1': '-01v.csv',
        '-0.4': '-04v.csv'
    }
    
    target_wl = 418.0  # Target wavelength (nm)
    header_skip = 9    # Skip header lines in CSV files

    # Storage for spectral data (Q1) and table data (Q2)
    all_spectra = []
    table_data = {}

    print("=" * 80)
    print("                    STARTING FILE PROCESSING")
    print("=" * 80)
    
    # --- 3. File Processing Loop ---
    for voltage_str, filename in files.items():
        if not os.path.exists(filename):
            print(f"⚠ Warning: File not found {filename}. Skipping this voltage.")
            continue
            
        try:
            df = pd.read_csv(filename, skiprows=header_skip)
            
            wl_col_name = df.columns[0]
            data_cols = [c for c in df.columns if 'Unnamed:' not in str(c)]
            eq_col_name = data_cols[-1]
            
            df = df.iloc[1:].copy()
            
            df[wl_col_name] = pd.to_numeric(df[wl_col_name], errors='coerce')
            df[eq_col_name] = pd.to_numeric(df[eq_col_name], errors='coerce')
            
            df = df.rename(columns={
                wl_col_name: 'Wavelength',
                eq_col_name: 'Absorbance' 
            })
            
            df_clean = df[['Wavelength', 'Absorbance']].dropna()
            
            closest_wl_index = (df_clean['Wavelength'] - target_wl).abs().idxmin()
            abs_at_418 = df_clean.loc[closest_wl_index, 'Absorbance']
            table_data[voltage_str] = abs_at_418
            
            df_clean['Voltage'] = voltage_str
            all_spectra.append(df_clean)
            
            print(f"✓ Successfully processed: {filename} (E={voltage_str} V)")

        except Exception as e:
            print(f"✗ Error: Failed to process {filename}. {e}")

    if not all_spectra:
        print("✗ Error: No data was processed. Check if CSV files are in the same folder.")
        return

    print("=" * 80)
    print("                    FILE PROCESSING COMPLETE")
    print("=" * 80)
    print()

    # --- 4. Q1: Generate Spectra Overlay Plot ---
    try:
        overlay_df = pd.concat(all_spectra)
        
        fig, ax = plt.subplots(figsize=(10, 6), facecolor='#fcfcfc')
        ax.set_facecolor('#f9fafb')
        
        # Modern muted palette with consistent brightness
        colors = sns.color_palette("rocket", n_colors=len(files))
        
        for i, voltage in enumerate(files.keys()):
            if voltage not in overlay_df['Voltage'].values:
                continue
            voltage_data = overlay_df[overlay_df['Voltage'] == voltage]
            ax.plot(
                voltage_data['Wavelength'],
                voltage_data['Absorbance'],
                label=f'{voltage} V',
                color=colors[i],
                linewidth=2.2,
                alpha=0.9,
            )
        
        ax.set_title('Q1 · Absorption Spectra at Equilibrium', fontsize=15, fontweight='bold', pad=18)
        ax.set_xlabel('Wavelength (nm)', fontsize=12, fontweight='bold')
        ax.set_ylabel('Absorbance', fontsize=12, fontweight='bold')
        ax.set_xlim(250, 500)
        
        legend = ax.legend(
            title="Applied Voltage",
            fontsize=10,
            title_fontsize=11,
            frameon=True,
            loc='upper right',
            edgecolor='#d2d6dc',
            facecolor='white'
        )
        legend.get_frame().set_linewidth(1)
        
        ax.grid(True)
        for spine in ['top', 'right']:
            ax.spines[spine].set_visible(False)
        for spine in ['left', 'bottom']:
            ax.spines[spine].set_color('#4b5563')
        ax.tick_params(axis='both', which='major', labelsize=10, colors='#374151')
        
        plt.tight_layout()
        plt.savefig('spectra_overlay_plot.png', dpi=600, bbox_inches='tight', facecolor=fig.get_facecolor())
        plt.show()
        print("✓ Q1 Plot saved: spectra_overlay_plot.png (600 DPI)\n")
    except Exception as e:
        print(f"✗ Error: Failed to generate Q1 plot. {e}\n")

    # --- 5. Q2 & Q4: Create Table and Nernst Plot ---
    try:
        table_df = pd.DataFrame.from_dict(table_data, orient='index', 
                                         columns=['Absorbance (at 418nm)'])
        table_df = table_df.reindex(files.keys())
        
        A_O = table_df.loc['+0.4', 'Absorbance (at 418nm)']
        A_R = table_df.loc['-0.4', 'Absorbance (at 418nm)']
        
        table_df['A-Ao'] = table_df['Absorbance (at 418nm)'] - A_O
        table_df['AR-A'] = A_R - table_df['Absorbance (at 418nm)']
        
        table_df['Ratio'] = table_df['A-Ao'] / table_df['AR-A']
        table_df['log(Ratio)'] = np.log10(table_df['Ratio'].replace([np.inf, -np.inf], np.nan))
        table_df['-log(Ratio)'] = -table_df['log(Ratio)']
        
        table_df['E(V)_numeric'] = [0.4, 0.2, 0.1, 0.0, -0.1, -0.4]
        table_df.index.name = 'Voltage (V)'
        
        # --- Q2: Print and Visualize Formatted Table ---
        print("=" * 80)
        print("                    Q2: DATA TABLE AT 418 nm")
        print("=" * 80)
        
        table_display = table_df.copy().round(6)
        print(table_display.to_string())
        print()
        
        fig, ax = plt.subplots(figsize=(13, 4.5), facecolor='#fcfcfc')
        ax.axis('tight')
        ax.axis('off')
        
        table_data_display = []
        for idx in table_df.index:
            row = [
                idx,
                f'{table_df.loc[idx, "Absorbance (at 418nm)"]:.6f}',
                f'{table_df.loc[idx, "A-Ao"]:.6f}',
                f'{table_df.loc[idx, "AR-A"]:.6f}',
                f'{table_df.loc[idx, "Ratio"]:.6f}',
                f'{table_df.loc[idx, "log(Ratio)"]:.6f}',
                f'{table_df.loc[idx, "-log(Ratio)"]:.6f}'
            ]
            table_data_display.append(row)
        
        col_labels = [
            'Voltage\n(V)',
            'Absorbance\n(at 418nm)',
            r'$A - A_O$',
            r'$A_R - A$',
            r'$\frac{A-A_O}{A_R-A}$',
            r'$\log(Ratio)$',
            r'$-\log(Ratio)$'
        ]
        
        table = ax.table(
            cellText=table_data_display,
            colLabels=col_labels,
            cellLoc='center',
            loc='center',
            colWidths=[0.12, 0.15, 0.13, 0.13, 0.15, 0.14, 0.14]
        )
        
        table.auto_set_font_size(False)
        table.set_fontsize(10)
        table.scale(1, 2)
        
        for i in range(len(col_labels)):
            cell = table[(0, i)]
            cell.set_facecolor('#111827')
            cell.set_text_props(weight='bold', color='white', fontsize=11)
            cell.set_height(0.12)
            cell.set_edgecolor('#111827')
        
        for i in range(1, len(table_data_display) + 1):
            for j in range(len(col_labels)):
                cell = table[(i, j)]
                cell.set_edgecolor('#e5e7eb')
                cell.set_linewidth(1)
                cell.set_facecolor('#f9fafb' if i % 2 == 0 else 'white')
        
        ax.text(
            0.5,
            0.98,
            'Q2 · Spectroelectrochemical Data (418 nm)',
            ha='center',
            va='top',
            transform=ax.transAxes,
            fontsize=14,
            fontweight='bold',
            color='#1f2937'
        )
        
        plt.tight_layout()
        plt.subplots_adjust(top=0.9)
        plt.savefig('data_table_418nm.png', dpi=600, bbox_inches='tight', facecolor=fig.get_facecolor())
        plt.show()
        print("✓ Q2 Table saved: data_table_418nm.png (600 DPI)\n")

        # --- Q4: Generate Nernst Plot ---
        plot_data = table_df.loc[(table_df.index != '+0.4') & (table_df.index != '-0.4')].copy()

        x = plot_data['-log(Ratio)']
        y = plot_data['E(V)_numeric']
        
        m, c = np.polyfit(x, y, 1)
        r_squared = np.corrcoef(x, y)[0, 1] ** 2
        
        fig, ax = plt.subplots(figsize=(10, 6), facecolor='#fcfcfc')
        ax.set_facecolor('#f9fafb')
        
        x_fit = np.linspace(x.min(), x.max(), 100)
        ax.plot(
            x_fit,
            m * x_fit + c,
            color='#2563eb',
            linestyle='-',
            linewidth=2.5,
            alpha=0.85,
            label='Linear Fit'
        )
        
        residuals = y - (m * x + c)
        std_residuals = np.std(residuals)
        ax.fill_between(
            x_fit,
            m * x_fit + c - 1.96 * std_residuals,
            m * x_fit + c + 1.96 * std_residuals,
            alpha=0.12,
            color='#93c5fd',
            label='95% CI'
        )
        
        ax.scatter(
            x,
            y,
            color='#ef4444',
            s=130,
            alpha=0.9,
            edgecolors='white',
            linewidths=2,
            label='Experimental Data'
        )
        
        ax.set_title('Q4 · Nernst Plot', fontsize=15, fontweight='bold', pad=18)
        ax.set_xlabel(r'$-\log\left(\frac{A-A_O}{A_R-A}\right)$', fontsize=12, fontweight='bold')
        ax.set_ylabel('E (V)', fontsize=12, fontweight='bold')
        ax.grid(True)
        for spine in ['top', 'right']:
            ax.spines[spine].set_visible(False)
        for spine in ['left', 'bottom']:
            ax.spines[spine].set_color('#4b5563')
        ax.tick_params(axis='both', which='major', labelsize=10, colors='#374151')
        
        textstr = f'$y = {m:.4f}x + {c:.4f}$\n$R^2 = {r_squared:.6f}$'
        props = dict(boxstyle='round,pad=0.55', facecolor='white', edgecolor='#d2d6dc', linewidth=1.25)
        ax.text(0.04, 0.95, textstr, transform=ax.transAxes, fontsize=11, va='top', bbox=props, color='#1f2937')
        
        ax.legend(loc='lower right', fontsize=10, frameon=True, edgecolor='#d2d6dc', facecolor='white')
        
        plt.tight_layout()
        plt.savefig('nernst_plot.png', dpi=600, bbox_inches='tight', facecolor=fig.get_facecolor())
        plt.show()
        print("✓ Q4 Plot saved: nernst_plot.png (600 DPI)\n")
        
        # --- 6. Analysis ---
        n = 0.059 / m
        print("=" * 80)
        print("                    ANALYSIS (FROM NERNST PLOT)")
        print("=" * 80)
        print(f"Slope (m): {m:.4f}")
        print(f"Intercept (Experimental E°): {c:.4f} V")
        print(f"R² (Coefficient of Determination): {r_squared:.6f}")
        print(f"Calculated electron transfer number (n = 0.059/m): {n:.4f}")
        print("=" * 80)
        
    except Exception as e:
        print(f"✗ Error: Failed to process Q2/Q4. {e}")

    print("\n✓ All tasks completed successfully!")


if __name__ == "__main__":
    process_spectro_data()
```
